### a. Write a Python NLTK program to find the number of male and female names in the names corpus.Print the first 10 male and female names.

#### Note: The names corpus contains a total of around 2943 male (male.txt) and 5001 female (female.txt) names.

```python
import nltk
nltk.download('all')
```

```python
nltk.corpus.names.fileids()
```

```python
maleNames = nltk.corpus.names.words('male.txt')[:10]
femaleNames = nltk.corpus.names.words('female.txt')[:10]
print("Male names: ", maleNames)
print("Female names: ", femaleNames)
```

### b. Write a Python NLTK program to print the first 15 random combine labeled male and labeled female names from names corpus.

```python
import random
```

```python
print("15 random male-female combined names:")
for i in range(15):
  a = random.randint(0, len(maleNames)-1)
  b = random.randint(0, len(femaleNames)-1)
  print(maleNames[a],"-",femaleNames[b])
```
